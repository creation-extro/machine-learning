```python
import pandas as pd
import numpy as np 

import sklearn
from sklearn.utils import shuffle 
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model , preprocessing
```

```python
data = pd.read_csv("car.data")
data.head()
```

```python
data = data.rename(columns={
    'vhigh': 'buying',
    'vhigh.1': 'maint',
    '2': 'door',
    '2.1': 'persons',
    'small': 'lu_boot',
    'low': 'safety',
    'unacc': 'class'
})
```

```python
data.head()
```

```python
# Create a LabelEncoder instance
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lu_boot = le.fit_transform(list(data["lu_boot"]))
safety	 = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))
print(buying)
```

```python
predict ="class"
```

```python
#For example, if you have two lists, x and y, 
#and you want to create pairs of corresponding elements, you can use zip:
a = [1, 2, 3]
b = [4, 5, 6]
zipped = list(zip(a, b))
print(zipped)
```

```python
x = list(zip(buying , maint , door , persons ,lu_boot , safety))
y = list(cls)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test ,y_train , y_test = train_test_split(x , y , test_size = 0.1)
print(x_train , y_test)
```

```python
model = KNeighborsClassifier(n_neighbors = 9)
model.fit(x_train , y_train )
acc = model.score(x_test , y_test)
print(acc)
```
